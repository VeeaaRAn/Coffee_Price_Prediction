```python
DS=DS.drop(columns=["Weekday","Month_name"])
```

```python
trainset,testset = train_test_split(DS,test_size=0.2)
```

```python
trainset
```

```python
trainset.drop(columns=["Date","Time"])
```

```python
targetcols=trainset["money"]
targetcols
```

```python
trainset=trainset.drop(columns=["money","Date", "Time"])
```

```python
trainset
```

```python
testset = testset.drop(columns=["money", "Date", "Time"])
```

```python
trainset = pd.get_dummies(trainset, columns=['coffee_name'])
```

```python
trainset
```

```python
trainset = pd.get_dummies(trainset, columns=['Time_of_Day'])
```

```python
trainset
```

```python
trainset.drop(columns=["Weekday_numeric","Month_numeric"])
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(trainset,targetcols)
```

```python
testset
```

```python
testset.drop(columns=["cash_type_card",	"Weekday_numeric",	"Month_numeric"])
```

```python
model.predict(testset)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Assuming 'testset_target' contains the actual money values for the test set
# You'll need to make sure you have stored the actual 'money' column for the test set before dropping it.
# Let's recreate the testset_target using the original DS dataframe
testset_target = DS.loc[testset.index, 'money']

predictions = model.predict(testset)

mae = mean_absolute_error(testset_target, predictions)
mse = mean_squared_error(testset_target, predictions)
r2 = r2_score(testset_target, predictions)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```
